```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import mne_bids

freq_bands = {
    'delta': (0.5, 4),
    'theta': (4, 8),
    'alpha': (8, 12),
    'beta': (12, 30),
    'gamma': (30, 100)
}

def compute_band_power(psd, freqs, band):
    band_freqs = np.logical_and(freqs >= band[0], freqs <= band[1])
    band_power = np.sum(psd[:, band_freqs], axis=1)
    return band_power

subjects = [f'{i:03d}' for i in range(1, 89)]
all_data = []
bids_root = './data/'
task = 'eyesclosed'
datatype = 'eeg'
suffix = 'eeg'

for subj in subjects:
    bids_path = BIDSPath(subject=subj, task=task, datatype=datatype, suffix=suffix, extension='.set', root=bids_root, check=False)
    raw = mne_bids.read_raw_bids(bids_path, verbose=True)
    psd, freqs = raw.compute_psd(method='welch', fmin=0.5, fmax=100, verbose=False).get_data(return_freqs=True)
    features = []
    for band, (fmin, fmax) in freq_bands.items():
        band_power = compute_band_power(psd, freqs, (fmin, fmax))
        features.append(band_power)
    subject_features = np.concatenate(features)
    all_data.append(subject_features)

X = np.array(all_data)
print(X.shape)

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1])
plt.title('t-SNE Plot of EEG Data (Using PSD)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

mean_point = np.mean(X_tsne, axis=0)
distances = np.linalg.norm(X_tsne - mean_point, axis=1)
threshold = np.mean(distances) + 3 * np.std(distances)
outlier_indices = np.where(distances > threshold)[0]
X_clean = np.delete(X, outlier_indices, axis=0)
print(X_clean.shape)

tsne_clean = TSNE(n_components=2, random_state=42)
X_tsne_clean = tsne_clean.fit_transform(X_clean)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne_clean[:, 0], X_tsne_clean[:, 1])
plt.title('t-SNE Plot of EEG Data (Outliers Removed)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()

print("Outlier indices:", outlier_indices)
```
