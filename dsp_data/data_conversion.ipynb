```python

import numpy as np
import mne_bids
from mne_bids import BIDSPath, read_raw_bids
import matplotlib.pyplot as plt

bids_root = './data/' 
subject = '001'   
task = 'eyesclosed'
datatype = 'eeg'
suffix = 'eeg'

bids_path = BIDSPath(subject=subject, task=task, datatype=datatype, suffix=suffix, extension='.set', root=bids_root, check=False)

raw = read_raw_bids(bids_path, verbose=True)

#plt.style.use('ggplot')

#psd = raw.compute_psd() 
# beräknar den spektrala effekttätheten (PSD) för den råa EEG-datan.
# PSD är ett mått på effekten som finns i signalen som en funktion av frekvens,
# vilket är användbart för att identifiera de dominerande frekvenserna i EEG-datan.

#psd.plot(picks='eeg',  average=False, color='blue', dB=True) 
#Average=False argumentet specificerar att vi inte vill ha medelvärde

#fig_raw = raw.plot(duration=5, n_channels=30, scalings='auto', title='Raw EEG Data')
# n_channels=30 argument specificerar antal kanaler som ska plottas

raw = raw.resample(100)
raw.plot_sensors(show_names=True)
```

```python
df = raw.to_data_frame()
df
```

```python
fp1_array = df['Fp1'].to_numpy()
fp2_array = df['Fp2'].to_numpy()
pz_array = df['Pz'].to_numpy()
time_array = df['time'].to_numpy()

print(fp1_array.shape)
training_data = np.column_stack((fp1_array, fp2_array, pz_array, time_array))
print(training_data.shape)
print(training_data)

fig, axs = plt.subplots(3, 1, figsize=(10, 8))

axs[0].plot(time_array, fp1_array)
axs[0].set_title('Fp1 Array')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Amplitude')

axs[1].plot(time_array, fp2_array)
axs[1].set_title('Fp2 Array')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Amplitude')

axs[2].plot(time_array, pz_array)
axs[2].set_title('Pz Array')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
